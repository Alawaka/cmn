```python
import numpy as np
```

```python
ann_file_pos = './positive_ims_sentences_chunks_train.txt'
ann_file_neg = './negative_outofclass_ims_sentences_chunks_train.txt'

save_file = './image_ann_dict.npz'
```

```python
def line2dict(l, label):
    entries = l.strip().split('\t')
    im_name = entries[0]
    explanation = entries[1]
    num_noun_phrase = int(entries[2])
    assert(num_noun_phrase == len(entries) - 3)
    
    noun_phrases = []
    for nph in entries[3:]:
        noun_phrases.append({'phrase': nph})
    
    return {'im_name': im_name, 'explanation': explanation, 'noun_phrases': noun_phrases, 'label': label}

def load_ann_file(fname, label):
    with open(fname) as f:
        lines = f.readlines()
    ann_list = []
    for l in lines:
        # skip a few (<10) annotations that have wrong line ending char
        try:
            l_dict = line2dict(l, label=label)
            ann_list.append(l_dict)
        except:
            print('error line:', l)
    return ann_list
```

```python
ann_list_pos = load_ann_file(ann_file_pos, label=True)
ann_list_neg = load_ann_file(ann_file_neg, label=False)

ann_list = ann_list_pos + ann_list_neg
```

```python
im_name2ann = {ann['im_name']: [] for ann in ann_list}
for ann in ann_list:
    im_name2ann[ann['im_name']].append(ann)
```

```python
np.savez(save_file, im_name2ann=im_name2ann)
```
